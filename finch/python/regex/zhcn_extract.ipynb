```
import re
```

```
"""
Suppose we want to extract location (such as 北京) from the sentence
"""

inp = "查 一 下 去 北 京 方 向 交 通 情 况 怎 么 样"
rule = r"(?:向|往|进|朝|去) (.*) (?:方 向)"
print(re.findall(rule, inp))
```

```
"""
Suppose we want to extract the slot name ADDRESS_NM from the sentence
"""

inp = "给 我 导航 到 <ADDRESS_NM> 上海市闸北区场中路270弄 </ADDRESS_NM> <ADDRESS_NM> 幸福新苑小区 </ADDRESS_NM> 的路线"
rule = r'<(.*?)>'
slot = re.findall(rule, inp)[0]
print(slot)
```

```
"""
Suppose now we want to take two address names out
We can use regex 'Look Behind':
(?<=r)d
matches d only if is preceded by r, but r will not be part of the overall regex match
"""

rule = r'(?<=<{}>) (.*?) (?=</.*>)'.format(slot)
print(re.findall(rule, inp))
```
