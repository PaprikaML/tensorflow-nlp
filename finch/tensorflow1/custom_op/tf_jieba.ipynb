```
import tensorflow as tf
from tf_jieba.jieba_op import jieba_cut

inp = tf.placeholder(dtype=tf.string)
out = jieba_cut(inp)

sess = tf.Session()
while True:
  raw_inp = input("Input: ").strip()
  if raw_inp == "q":
      break
  res = sess.run(out, feed_dict={inp: raw_inp})
  print("Output: {}".format(res.decode('utf-8')))
sess.close()
```
